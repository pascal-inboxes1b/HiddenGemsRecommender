# Hidden Gems Travel Recommender (Europe Hotels)
Uses the 515K Europe hotel reviews dataset with clean location + NLP features.

```
import pandas as pd
import numpy as np
import re
from textblob import TextBlob
```

## Load Dataset

```
df = pd.read_csv('Hotel_Reviews.csv')

# Optional: sample for speed
df = df.sample(50000, random_state=42)

df.head()
```

## Basic Cleaning

```
df['review_text'] = df['Positive_Review'] + ' ' + df['Negative_Review']

df = df[['Hotel_Name','Hotel_Address','Reviewer_Score','review_text',
         'Total_Number_of_Reviews','lat','lng']]

df.dropna(subset=['review_text'], inplace=True)
```

## Extract City from Address

```
def extract_city(address):
    parts = address.split()
    if len(parts) > 2:
        return parts[-2]
    return None

df['city'] = df['Hotel_Address'].apply(extract_city)
```

## NLP: Sentiment

```
df['sentiment'] = df['review_text'].apply(
    lambda x: TextBlob(str(x)).sentiment.polarity
)
```

## NLP: Quiet vs Crowded

```
QUIET = ["quiet","peaceful","calm","relaxing","tranquil","secluded"]
CROWDED = ["crowded","busy","noisy","packed","touristy"]

def has_negation(text, keyword):
    return re.search(rf"(not|no|never)\s+\w*\s*{keyword}", text) is not None

def score(text):
    text = str(text).lower()
    quiet = any(q in text for q in QUIET)
    crowded = any(c in text and not has_negation(text,c) for c in CROWDED)
    return int(quiet), int(crowded)

df[['quiet_flag','crowded_flag']] = df['review_text'].apply(
    lambda x: pd.Series(score(x))
)
```

## Aggregate at Hotel Level

```
hotel_stats = df.groupby(['Hotel_Name','city','lat','lng']).agg(
    avg_rating=('Reviewer_Score','mean'),
    review_count=('Reviewer_Score','count'),
    sentiment=('sentiment','mean'),
    quiet_score=('quiet_flag','mean'),
    crowded_score=('crowded_flag','mean')
).reset_index()
```

## Hidden Gem Score

```
hotel_stats['hidden_score'] = hotel_stats['avg_rating'] * (
    1 / np.log(1 + hotel_stats['review_count'])
)
```

## Recommendation Function

```
def recommend(df, avoid_crowds=1.0, top_k=5):
    score = (
        df['hidden_score']
        + df['quiet_score'] * avoid_crowds
        - df['crowded_score'] * avoid_crowds
    )
    return df.sort_values(by=score, ascending=False).head(top_k)

recommend(hotel_stats)
```
