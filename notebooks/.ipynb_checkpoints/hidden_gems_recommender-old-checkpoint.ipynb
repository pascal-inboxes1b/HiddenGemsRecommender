```python
cities = pd.read_csv('../data/worldcities.csv')
#reviews = pd.read_csv('../data/extracted/reviews.csv')
cities.head()
```

## Filter European Cities

```python
europe_countries = ["France","Italy","Spain","Germany","Netherlands","Portugal","Greece","Switzerland","Austria","Belgium","Denmark","Sweden","Norway","Finland","Poland","Czechia","Hungary","Ireland"]

cities = cities[cities["country"].isin(europe_countries)]
#cities = cities[["city","country","lat","lng","population"]]
cities = cities[["city","country","lat","lon","population"]]
cities = cities[cities["population"] > 100000]

cities["city_clean"] = cities["city"].str.lower()
city_list = cities["city_clean"].unique().tolist()
```

## City Matching

```python
def exact_city_match(text, city_list):
    text = text.lower()
    for city in city_list:
        if re.search(rf"\b{re.escape(city)}\b", text):
            return city
    return None

def fuzzy_city_match(text, city_list, threshold=90):
    text = text.lower()
    match, score, _ = process.extractOne(text, city_list, scorer=fuzz.partial_ratio)
    if score >= threshold:
        return match
    return None

def match_city(text, city_list):
    if pd.isna(text):
        return None
    city = exact_city_match(text, city_list)
    if city:
        return city
    return fuzzy_city_match(text, city_list)
```

## Quiet vs Crowded NLP Features

```python
QUIET_KEYWORDS = ["quiet","peaceful","calm","serene","relaxing","tranquil","uncrowded","secluded"]
CROWDED_KEYWORDS = ["crowded","busy","packed","touristy","overcrowded","noisy","loud","queue"]

def has_negation(text, keyword):
    return re.search(rf"(not|no|never|hardly)\s+\w*\s*{keyword}", text) is not None

def score_review(text):
    text = str(text).lower()
    quiet = any(kw in text for kw in QUIET_KEYWORDS)
    crowded = any(kw in text and not has_negation(text, kw) for kw in CROWDED_KEYWORDS)
    return int(quiet), int(crowded)

reviews[["quiet_flag","crowded_flag"]] = reviews["Review"].apply(lambda x: pd.Series(score_review(x)))
```

## Sentiment

```python
reviews["sentiment"] = reviews["Review"].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

## Aggregate by City

```python
city_stats = reviews.dropna(subset=["city"]).groupby("city").agg(
    avg_rating=("Rating","mean"),
    review_count=("Rating","count"),
    sentiment_score=("sentiment","mean"),
    quiet_score=("quiet_flag","mean"),
    crowded_score=("crowded_flag","mean")
).reset_index()
```

## Merge with Cities

```python
cities = cities.merge(city_stats, left_on="city_clean", right_on="city", how="left")

global_avg = reviews["Rating"].mean()
global_sent = reviews["sentiment"].mean()

cities["avg_rating"].fillna(global_avg, inplace=True)
cities["review_count"].fillna(1, inplace=True)
cities["sentiment_score"].fillna(global_sent, inplace=True)
cities["quiet_score"].fillna(0, inplace=True)
cities["crowded_score"].fillna(0, inplace=True)
```

## Hidden Gem Score

```python
cities["hidden_score"] = cities["avg_rating"] * (1 / np.log(1 + cities["review_count"]))
```

## Recommendation Function

```python
def recommend(df, user_pref, top_k=5):
    base = (
        user_pref["beach"] * np.random.rand(len(df)) +
        user_pref["culture"] * np.random.rand(len(df)) +
        user_pref["food"] * np.random.rand(len(df))
    )

    score = base * df["hidden_score"] + df["quiet_score"] * user_pref["avoid_crowds"] - df["crowded_score"] * user_pref["avoid_crowds"]

    return df.sort_values(by=score, ascending=False).head(top_k)

user_pref = {"beach":0.8,"culture":0.5,"food":0.6,"avoid_crowds":1.0}

recommend(cities, user_pref)
```
