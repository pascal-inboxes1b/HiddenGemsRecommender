```python
# Province_Name is actually City so renaming that column
df.rename(columns={'Province_Name': 'City'}, inplace=True)

# Remove Trailing Whitespace
df["City"] = df["City"].str.strip()
df['City'].unique().tolist()
```

```python
# Create Review Text
df["review_text"] = (
    df["Positive_Review"].fillna("") + " " +
    df["Negative_Review"].fillna("")
)
# Keep Only Needed Columns

df = df[[
    "Hotel_Name",
    "Hotel_Address",
    "Reviewer_Score",
    "Total_Number_of_Reviews",
    "review_text",
    "lat",
    "lng"
]]

df.dropna(subset=["review_text", "lat", "lng"], inplace=True)
#print(df.loc[0, 'Hotel_Address']) 
```

## City Extraction

```python
def extract_city(address):
    if pd.isna(address):
        return None
    
    parts = [p.strip() for p in address.split(" ")]
    
    # Typical format: street, city, country
    if len(parts) >= 3:
        return parts[-2]
    
    if len(parts) == 2:
        return parts[0]
    
    return None

df["city"] = df["Hotel_Address"].apply(extract_city)
```

```python
df["city"].unique().tolist()
```

## Sentiment Analysis

```python
df["sentiment"] = df["review_text"].apply(
    lambda x: TextBlob(str(x)).sentiment.polarity
)

#df["sentiment"].head()
```

## Quiet vs Crowded NLP Features

```python
QUIET_KEYWORDS = [
    "quiet", "peaceful", "calm", "serene",
    "relaxing", "tranquil", "secluded", "not busy"
]

CROWDED_KEYWORDS = [
    "crowded", "busy", "packed", "touristy",
    "overcrowded", "noisy", "loud", "queue"
]

def has_negation(text, keyword):
    pattern = rf"(not|no|never|hardly)\s+\w*\s*{keyword}"
    return re.search(pattern, text) is not None


def score_review(text):
    text = str(text).lower()
    
    quiet = any(q in text for q in QUIET_KEYWORDS)
    
    crowded = False
    for c in CROWDED_KEYWORDS:
        if c in text and not has_negation(text, c):
            crowded = True
    
    return int(quiet), int(crowded)


df[["quiet_flag", "crowded_flag"]] = df["review_text"].apply(
    lambda x: pd.Series(score_review(x))
)
```

## Aggregate to Hotel Level

```python
hotel_stats = df.groupby(
    ["Hotel_Name", "city", "lat", "lng"]
).agg(
    avg_rating=("Reviewer_Score", "mean"),
    review_count=("Reviewer_Score", "count"),
    sentiment_score=("sentiment", "mean"),
    quiet_score=("quiet_flag", "mean"),
    crowded_score=("crowded_flag", "mean")
).reset_index()
```

## Recommendation Function

```python
def recommend(df, user_pref, top_k=5):
    base = (
        user_pref["beach"] * np.random.rand(len(df)) +
        user_pref["culture"] * np.random.rand(len(df)) +
        user_pref["food"] * np.random.rand(len(df))
    )

    score = base * df["hidden_score"] + df["quiet_score"] * user_pref["avoid_crowds"] - df["crowded_score"] * user_pref["avoid_crowds"]

    return df.sort_values(by=score, ascending=False).head(top_k)

user_pref = {"beach":0.8,"culture":0.5,"food":0.6,"avoid_crowds":1.0}

recommend(cities, user_pref)
```
